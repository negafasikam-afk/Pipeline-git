# setting a parameter called files 

```python
files = 'orders'
```

# Reading the sales csv file from lakehouse

```python
df = spark.read.format("csv").option("header", "true").load(f"Files/new_data/{files}")

display(df)
```

# Writing/saving the csv file as delta table in lakehouse

```python
df.write.mode("overwrite").format("delta").saveAsTable(f"dbo.{files}")
```

```python
df_new = spark.read.table(f"dbo.{files}")
display(df_new)
```
